```python
%matplotlib inline

import pandas as pd

from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")
```

## Pandas DataFrame

We can render Pandas DataFrames natively:

```python
data_age = [1, 5, 10, 15, 18]
data_height = [76, 120, 140, 169, 176]

data = pd.DataFrame({"Age": data_age, "Height": data_height})
data
```

## Matplotlib Plots

We can render Matplotlib Plots natively in SVG:

```python
ax = data.plot.scatter(x="Age", y="Height")
_ = ax.axline(data.iloc[0], data.iloc[-1], color="red", linestyle="dashed")
```
